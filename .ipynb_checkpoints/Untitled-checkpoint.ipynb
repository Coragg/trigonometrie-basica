# Mini biblioteca de matematica

```julia
function even_or_odd(number)
    if number % 2 == 0
        return true
    else
        return false
    end
end
```

```julia
function factorial(number)
    if number <= 1
        return 1
    end
    fact = 1
    for iteration in 1:number 
        fact *= iteration
    end
    return fact
end
```

```julia
function aproximation_pi_fermat(n)
    iteration = 1
    serie_value = 1
    while n > iteration
        if even_or_odd(iteration) == false
            serie_value -= 1/(2 * iteration + 1)
        else
            serie_value += 1/(2 * iteration + 1)
        end
        iteration += 1
    end    
    return 4 * serie_value
end
```

```julia
function coseno(x, n = 10)
    calculete = 0 
    iteration = 0
    while n > iteration
        calculete += ((-1)^iteration)*((x^(2*iteration))/(factorial(2*iteration)))
        iteration += 1
    end
    return calculete  
end
```

```julia
function sino(x, n = 10)
    caculete = 0
    iteration = 0
    while n > iteration
        caculete += ((-1)^iteration)*((x^(2*iteration + 1))/factorial(2*iteration + 1))
        iteration += 1
    end
    return caculete    
end
```

```julia
function limite_euler(n)
    return (1 + (1/n))^n
end
```

```julia
function tangent(x, n = 10)
    return sino(x, n) / coseno(x, n)
end
```

```julia
function arcsin(x, n = 10)
   return (sino(x, n)) ^ -1 
end
```

```julia
function arccos(x, n = 10)
   return (coseno(x, n)) ^ -1 
end
```

```julia
function arctan(x, n=10)
    return (tangent(x, n)) ^ -1
end
```

```julia
function raices(radical, indice::Int64 = 2)
    if indice != 0
        return radical ^ (1/indice)
    else
        return NaN
    end
end
```

```julia
function potencia(base::Int64, exponent::Int64 = 2)
    return base ^ exponent
end    
```

## Proces to create graphics

```julia
using Pkg
Pkg.add("Plots")
```

```julia
using Plots
```

```julia
numbers = [0 , 90, 180, 270,360]
data_sin = []
data_cos = []
```

```julia
for i in numbers
    push!(data_sin, sino(degrees_to_radian(i)))
    push!(data_cos, coseno(degrees_to_radian(i))) 
end
```

```julia
numbers # show data of the vector
```

```julia
gr()
plot(numbers, data_cos, label="coseno", title = "grafica coseno")
scatter!(numbers, data_cos, label="points")
```

```julia
plot(numbers, data_sin, label="seno", title = "grafica de seno")
```

```julia
plot(numbers, [data_cos data_sin], label=["coseno" "seno"], title = "grafica de seno y coseno")
scatter!(numbers, [data_cos data_sin], label=["points coseno" "points seno"])
ylabel!("Y")
xlabel!("Grados")
```
